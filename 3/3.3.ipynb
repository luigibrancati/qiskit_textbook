# Superdense Coding

```python
# Do the necessary imports
import numpy as np
from qiskit import *
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from qiskit.extensions import Initialize
from qiskit_textbook.tools import array_to_latex
```

The process is loosely as follow: A and B share an entangled pair of qubit in a Bell state. Then, A can perform a series of gates on his qubit so as to modify the shared entangled state (the Bell state is converted into another Bell state) and then can send his state to B. B can then measure the state and extract the two classical bits (from measuring the relative phase and parity).

## Step 1

C prepares two qubits in a Bell state. As seen before, he applies Hadamard gate and the a CNOT, starting from $|00>$.

```python
def charlie_gates(qc, a, b):
    qc.barrier()
    qc.h(a)
    qc.cx(a,b)
```

## Step 2

C sends the two qubits to A and B, one each. A now needs to encode the two bits she wants to send in the state A and B are sharing. She decides an encoding like so:

In the encoding rules above are shown the gates A needs to apply to her qubit in order to obtain the desired state and thus the desired 2 bit message. Notice that the first bit is 1 if the two qubit would give opposite outcomes when measured (parity), while the second bit is 1 when we have a negative relative phase. These 4 states correspond to the 4 Bell basis states for a 2 qubit system. 

Then A just needs to send her qubit to B.

```python
def alice_gates(msg, qc, a):
    qc.barrier()
    if len(msg)==2:
        if msg[0]=='1':
            qc.x(a)
        if msg[1]=='1':
            qc.z(a)
    else:
        print("Invalid message. Sending '00'")
```

## Step 3

Bob receives the two bits and is now able to recover the classical 2 bits. The best way to do so is to invert the initial procedure C did to entangle the two qubits. Each Bell state can be created by using C procedure from one of the four standard basis states $|00>, |01>, |10>, |11>$, and thus inverting it will return the correspong state, which corresponds to the 2 bit message.

B then applies a CNOT and then an Hadamard on the first (A) qubit, and finally measure its qubits:


After the measurement B will get with certainty the message A sent.

```python
def bob_gates(qc, a, b):
    qc.barrier()
    qc.cx(a,b)
    qc.h(a)
```

## Simulating the protocol

```python
qc = QuantumCircuit(2)

msg = "11"

charlie_gates(qc, 0, 1)
alice_gates(msg, qc, 0)
bob_gates(qc, 0, 1)

qc.measure_all()

qc.draw()
```

```python
backend = Aer.get_backend('qasm_simulator')
```

```python
job_sim = execute(qc, backend, shots=1024)
sim_result = job_sim.result()
measurement_result = sim_result.get_counts(qc)
print(measurement_result)
plot_histogram(measurement_result)
```

The simulator give the correct message 100% of the times.

## Real Hardware

```python
from qiskit import IBMQ
from qiskit.providers.ibmq import least_busy
shots = 512
```

```python
IBMQ.load_account()
provider = IBMQ.get_provider(hub='ibm-q', group='open', project='main')
```

```python
from qiskit.tools.monitor import job_monitor

IBMQ_backend = least_busy(
    provider.backends(
        filters=lambda b: (b.configuration().n_qubits>=2 
                           and not b.configuration().simulator 
                           and b.status().operational==True)))
print("Least busy: ", IBMQ_backend)

job = execute(qc, IBMQ_backend, shots=shots)
job_monitor(job)
```

```python
result = job.result()
plot_histogram(result.get_counts(qc))
```

```python
print(result.get_counts())
correct = result.get_counts(qc)[msg]
accuracy = correct/shots*100.
print("Accuracy = %.2f%%" % accuracy)
```
