```python
# Do the necessary imports
import numpy as np
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, execute, BasicAer, IBMQ
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from qiskit.extensions import Initialize
from qiskit_textbook.tools import random_state, array_to_latex
```

```python
# The protocol uses 3 qubits and 2 bits

qr = QuantumRegister(3, name='q')
crz = ClassicalRegister(1, name='crz')
crx = ClassicalRegister(1, name='crx')
teleportation_circuit = QuantumCircuit(qr, crz, crx)
teleportation_circuit.draw()
```

### Step 1

A and B need to share a Bell state. We'll use the qubits q1 and q2.

To make the Bell + state (we only consider q1 and q2 here):

$$
\begin{aligned}
&1_2 \otimes H |00> \rightarrow |0+>\\
&CNOT(1,2)|0+> \rightarrow \frac{1}{\sqrt{2}} (|00>+|11>)
\end{aligned}
$$

```python
def create_bell_pair(qc, a, b):
    """Creates bell pair between qubits a and b"""
    qc.h(a) #|+>
    qc.cx(a,b)
```

```python
# The protocol uses 3 qubits and 2 bits

qr = QuantumRegister(3, name='q')
crz, crx = ClassicalRegister(1, name='crz'), ClassicalRegister(1, name='crx')
teleportation_circuit = QuantumCircuit(qr, crz, crx)
teleportation_circuit.draw()

create_bell_pair(teleportation_circuit, 1, 2)
teleportation_circuit.draw()
```

A owns q1 and B q2, and they are far away. A is trying to send q0, which is now in |0>

### Step 2

A now applies a CNOT to q0 and q1 conditioned on q0, and then an hadamard to q1.

```python
def alice_gates(qc, psi, a):
    qc.barrier()
    qc.cx(psi, a)
    qc.h(psi)
```

```python
# The protocol uses 3 qubits and 2 bits

qr = QuantumRegister(3, name='q')
crz, crx = ClassicalRegister(1, name='crz'), ClassicalRegister(1, name='crx')
teleportation_circuit = QuantumCircuit(qr, crz, crx)
teleportation_circuit.draw()

create_bell_pair(teleportation_circuit, 1, 2)
alice_gates(teleportation_circuit, 0, 1)
teleportation_circuit.draw()
```

### Step 3

A now measures both q0 and q1 and sends the two bits to B.

```python
def measure_and_send(qc, a, b):
    qc.barrier()
    qc.measure(a,0)
    qc.measure(b,1)
```

```python
# The protocol uses 3 qubits and 2 bits

qr = QuantumRegister(3, name='q')
crz, crx = ClassicalRegister(1, name='crz'), ClassicalRegister(1, name='crx')
teleportation_circuit = QuantumCircuit(qr, crz, crx)
teleportation_circuit.draw()

create_bell_pair(teleportation_circuit, 1, 2)
alice_gates(teleportation_circuit, 0, 1)
measure_and_send(teleportation_circuit,0,1)
teleportation_circuit.draw()
```

### Step 4

B has qubit q2 and has to follow the following scheme, based on the value of the two bits received by A:
- 00 $\rightarrow$ Do nothing
- 01 $\rightarrow$ Apply X
- 10 $\rightarrow$ Apply Z
- 11 $\rightarrow$ Apply ZX

```python
def bob_gates(qc, qubit, crz, crx):
    qc.barrier()
    #we use a c_if to control the quantum gates with a classical bit
    qc.x(qubit).c_if(crx,1) # applies the gate if bit is 1
    qc.z(qubit).c_if(crz,1) # applies the gate if bit is 1    
```

```python
# The protocol uses 3 qubits and 2 bits

qr = QuantumRegister(3, name='q')
crz, crx = ClassicalRegister(1, name='crz'), ClassicalRegister(1, name='crx')
teleportation_circuit = QuantumCircuit(qr, crz, crx)
teleportation_circuit.draw()

create_bell_pair(teleportation_circuit, 1, 2)
alice_gates(teleportation_circuit, 0, 1)
measure_and_send(teleportation_circuit,0,1)
bob_gates(teleportation_circuit, 2, crz, crx)
teleportation_circuit.draw()
```

# Simulating the protocol

We now test it works. We first initialize A qubit q0 in a random state $|\psi>$. We use the function random_state to generate a random qubit state, and then intialize q0 in this state with Initialize.

```python
psi = random_state(1)
array_to_latex(psi, pretext="|\\psi\\rangle = ")
psi
```

```python
init_gate = Initialize(psi)
init_gate.label = "init"
```

```python
init_circuit = QuantumCircuit(qr,crz,crx)
init_circuit.append(init_gate, [0])
```

We use the statevector simulator in order to check whether the qubit q2 is in this state at the end of the protocol.

```python
statevector_backend = BasicAer.get_backend('statevector_simulator')
```

The full circuit with initialization is

```python
qc_teleport = init_circuit + teleportation_circuit
qc_teleport.draw()
```

In order to use plot_bloch_multivector to compare the states, we need to build a 2 qubit circuit with the initial state.

```python
qc_tmp = QuantumCircuit(2)
qc_tmp.append(init_gate, [0])
out_state = execute(qc_tmp, statevector_backend).result().get_statevector()
plot_bloch_multivector(out_state)
```

```python
out_vector = execute(qc_teleport, statevector_backend).result().get_statevector()
plot_bloch_multivector(out_vector)
```

```python
out_vector
```

With the help of the bloch sphere we see the initial state (q0) is the same as the final state of q2.

## QASM SIMULATOR

The QASM simulator simulates the working of an actual quantum computer, so we won't be able now to directly compare two quantum states as we did above.

We initialize as before, the initialization resets the qubit to $|0>$ and the turns it into $|\psi>$

```python
inverse_init_gate = init_gate.gates_to_uncompute()
```

So now we have the gates that perform the operation

$$
|\psi> \longrightarrow |0>
$$

So if the algorithm worked, we have to measure the 0 state with certainty. We need to just add it to the circuit and then measure the state multiple times to sample it's probability

```python
inverse_init_circuit = QuantumCircuit(qr, crx, crz)
inverse_init_circuit.append(inverse_init_gate, [2])
```

```python
qc_teleport_qasm = qc_teleport + inverse_init_circuit
qc_teleport_qasm.draw()
```

We add a bit to measure the third qubit

```python
cr_result = ClassicalRegister(1)
qc_teleport_qasm.add_register(cr_result)
qc_teleport_qasm.measure(2,2)
qc_teleport_qasm.draw()
```

```python
qasm_backend = BasicAer.get_backend('qasm_simulator')
counts = execute(qc_teleport_qasm, qasm_backend, shots=20000).result().get_counts()
plot_histogram(counts)
```

The q2 qubit (leftmost bit) always returns 0, so this means the protocol worked.

# Teleportation on a real quantum computer

On real hardware, we cannot make operations after measurements, so the circuit above won't work.

Deferred measurement principle: in any algorithm, all measurements can be postponed until the end.

So we can replace a classically controlled gate with an equivalent controlled-gate. For example, the X operation on q2 controlled on the bit measured on q1 can be represented as a CNOT on q2 controlled on q1. And then measure at the very end.

Notice though that in this way the advantage of sending a quantum state by just sending 2 classical bits is lost.

Let's rewrite bob gates then:

```python
def new_bob_gates(qc, a, b, c):
    qc.barrier()
    qc.cx(b,c)
    qc.cz(a,c)
```

```python
teleportation_circuit_IBMQ = QuantumCircuit(3, 1)
```

```python
create_bell_pair(teleportation_circuit_IBMQ, 1, 2)
alice_gates(teleportation_circuit_IBMQ, 0, 1)
new_bob_gates(teleportation_circuit_IBMQ, 0, 1, 2)
teleportation_circuit.draw()
```

```python
qc_teleport_IBMQ = init_circuit + teleportation_circuit_IBMQ + inverse_init_circuit
qc_teleport_IBMQ.measure(2,0)
qc_teleport_IBMQ.draw()
```

```python
IBMQ.load_account()
provider = IBMQ.get_provider(hub='ibm-q', group='open', project='main') 
```

```python
from qiskit.providers.ibmq import least_busy
IBMQ_backend = least_busy(provider.backends(filters=lambda b: b.configuration().n_qubits>=3 and not b.configuration().simulator and b.status().operational==True))
print(IBMQ_backend)
```

```python
job_exp = execute(teleportation_circuit_IBMQ, IBMQ_backend, shots=8192)
```

```python
from qiskit.tools.monitor import job_monitor
job_monitor(job_exp)

exp_result = job_exp.result()
exp_measurement_result = exp_result.get_counts(teleportation_circuit_IBMQ)
print(exp_measurement_result)
plot_histogram(exp_measurement_result)
```

```python
error_rate_percent = (
    sum([exp_measurement_result[result] for result in exp_measurement_result.keys() if result[0]=='1'])/
                          sum(list(exp_measurement_result.values())))*100.
print("The experimental error rate : ", error_rate_percent, "%")
```
