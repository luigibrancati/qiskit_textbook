# Proving Unversality

Most of the chapter was theory. This is the last section.

We just noticed we can achieve this unitary

U = exp{i(a*X + b*Z)}

by using Rx(2a)=exp{i a*X} and Rz=exp{ib*Z} thanks to the Trotter formula

U = lim (exp{i(a*X)/n} * exp{i(b*Z)/n})^n

This has the advantage it can be used for much more complex unitaries on 
multiple qubits, like

U = exp{i(a X x X x X + b Z x Z x Z)}

We can make the x side by just using Rx and CNOTS

```python
theta=pi
qc = QuantumCircuit(3)
qc.cx(0,2)
qc.cx(0,1)
qc.rx(theta,0)
qc.cx(0,1)
qc.cx(0,2)
qc.draw()
```

Add a few Hadamard can be added to the above to make the Zs side

```python
theta=pi
qc = QuantumCircuit(3)
for i in range(3):
    qc.h(i)
qc.cx(0,2)
qc.cx(0,1)
qc.rx(theta,0)
qc.cx(0,1)
qc.cx(0,2)
for i in range(3):
    qc.h(2-i)
qc.draw()
```

and thus we can use the Trotter formula to reproduce the same approximations 
on multiple qubits.
